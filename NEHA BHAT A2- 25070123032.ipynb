```python
# Fibonacci Series using while loop

num = int(input("Enter the number of iterations: "))
n1 = 0
n2 = 1
print(n1, n2 , end = ' ')
num = num - 2

while(num > 0):
  n3 = n1 + n2
  print(n3, end = ' ')
  n1 = n2
  n2 = n3
  num -= 1
```

```python
# Reverse a Number

num = int(input("Enter the number to be reversed: "))
reverse = 0

while (num > 0):
  remainder = num % 10
  reverse = (reverse * 10) + remainder
  num = num // 10
print(reverse)
```

```python
# Check Palindrome or not

text = input("Enter the String: ")
text = text.lower()
text1 = text[::-1]

if (text == text1):
  print("It is a Palindrome")
else:
  print("It is not a Palindrome")
```

```python
# Check Palindrome or not using While loop

text = input("Enter the String: ")
text = text.lower()
result = True
i = 0
j = len(text) - 1

while(i < j):

  if(text[i] != text[j]):
    result = False
    break
  else:
     i += 1
     j -=1

if(result):
  print("It is a Palindrome")
else:
  print("It is not a Palindrome")
```

```python
# SEARCH AN ELEMENT IN A LIST

s = [176, 175, 174, 166, 142, 141, 31, 32, 33, 29, 28, 27]

prn = int(input("Enter a 3 digit PRN: "))
i = 0
found = False

while(i < len(s)):
  if(prn == s[i]):
    print("Element found at index ", i + 1)
    found = True
    break
  i += 1

if not found:
  print("Element not Found")

```
